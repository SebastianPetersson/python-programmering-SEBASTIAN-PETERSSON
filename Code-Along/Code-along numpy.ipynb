```python
import numpy as np
import matplotlib.pyplot as plt
n = 10

def roll_ndice(n):
    rng = np.random.default_rng()
    rng.integers(1, 6, n, endpoint=True)
    return n

mean = list()
for n in [10**p for p in range(1,9)]:
    mean += [np.mean(roll_ndice(n))]

x = list(range(len(mean)))
plt.plot(x, mean)





```
