```python
class Robot:
    def __init__(self, n, f_c, w):
        self.name = n
        self.color = f_c
        self.weight = w

    def introduce_yourself(self):
        print(f"Hi, my name is {self.name}, I weigh {self.weight} and my favorite color is {self.color}.")


robot1 = Robot('Wall_E', 'beige', '240 kg')
robot2 = Robot('Eve', 'pink', '545 kg')

class Person:
    def __init__(self, n, p, i):
        self.name = n
        self.personality = p
        self.is_sitting = i

    def sit_down(self):
        self.is_sitting = True

    def stand_up(self):
        self.is_sitting = False
    
p1 = Person('Alice', 'shy', False)
p2 = Person('Becky', 'Talkative', True)

p1.robot_owned = robot2
p2.robot_owned = robot1

p1.robot_owned.introduce_yourself()

p1.stand_up()
p1.sit_down()

if p1.is_sitting == True:
    print('Alice is sitting down.')
else:
    print('Alice is standing up.')
```

```python
class BankAccount:
    def __init__(self, owner):
        self.owner = owner
        self._balance = 0
    
    def show_balance(self):
        """Shows the current balance on your personal account."""
        print(f"Saldo: {self._balance} kr.")
    
    def deposit(self, amount):
        """Lets the user deposit money into a bank account."""
        if amount <= 0:
            print(f"Du kan inte sätta in negativa pengar... nu försökte du sätta in {amount} kr vilket absolut inte går.")
        else:
            ("{amount} sattes in på ditt konto.")
            self._balance += amount
            print(f"Du satte in {amount} kr.")

    def withdraw(self, amount):
        """Lets the user withdraw money from a bank account."""
        if amount <= 0:
            print(f"Du kan inte ta ut negativa pengar... nu försökte du ta ut {amount} kr vilket absolut inte går.")
        elif amount > self._balance:
            print(f"Du har inte nog med pengar på kontot, ditt saldo är {self._balance}.")
        else:
            self._balance -= amount
            print(f"Du tog ut {amount} kr, var vänlig ta pengarna.")

my_acc = BankAccount('Sebastian')

my_acc.show_balance()
my_acc.deposit(200)
my_acc.show_balance()
my_acc.withdraw(35)
my_acc.show_balance()
my_acc.withdraw(170)
my_acc.withdraw(165)
my_acc.show_balance()
my_acc.deposit(1000000)
print("Gott e d.")





        
```

```python
class Car:

    def __init__(self, brand, fuel, consumption, distance):
    
        self.brand = brand
        self.fuel = fuel
        self.consumption = consumption
        self.distance = distance
    
    def drive(self, miles):
        if self.fuel - (miles * self.consumption) < 0:
            print("There's not enough fuel in the tank.")
        else:
            self.fuel -= (miles * self.consumption)
            self.distance += miles
            print(f'You drove {miles} miles with your {self.brand}. That means you used {miles * self.consumption:.2f} liters of fuel.')

    def refuel(self, liters):
        if self.fuel == 55:
            print(f"Your tank is already full.")
        else:
            if self.fuel + liters <= 55:
                self.fuel += liters
                print(f"You refueled your car {self.brand} with {liters} liters of gasoline, you now have {self.fuel} liters in your tank.")
            else:
                print(f"The maximum amount you can refuel is {55 - self.fuel} liters.")
    def status(self):
        if self.fuel == 0.00:
            print(f"Your car {self.brand} currently has driven {self.distance} miles. You have soppatorsk.")
        else:
            print(f"Your car {self.brand} currently has driven {self.distance} miles. There is {self.fuel:.2f} liters left in the tank.")
        

Kia = Car('Kia Picanto', 45, 0.6, 0)


Kia.status()
Kia.drive(50)
Kia.drive(20)
Kia.status()
Kia.drive(4)
Kia.status()
Kia.drive(1)
Kia.status()
Kia.refuel(20)


```

```python
class Dog:
    def __init__(self, name, gender, age, energy):
        self.name = name
        self.gender = gender
        self.age = age
        self.energy = energy

    def bark(self):
        print(f"Your dog {self.name} barks: 'BARK!'")

    def birthday(self):
        self.age += 1
        print(f"It's {self.name}'s birthday today! {self.gender} age is now {self.age}!  ")

    def sleep(self, hours):

        if self.energy + (hours * 15) >= 100:
            self.energy = 100
            print(f"{self.name} Slept for {hours} hours and he now has {self.energy} % energy.")
            
        else:
            self.energy += hours * 15
            print(f"{self.name} slept for {hours} hours, now {self.name} has {self.energy} % energy.")

    def play(self, minutes):

        energy_loss = minutes * 2
        if self.energy - energy_loss <= 0:
            self.energy = 0
            print(f"{self.name} played for {minutes} minutes, now {self.name} has {self.energy} % energy left.")
        else: 
            self.energy -= energy_loss
            print(f"{self.name} played for {minutes} minutes, now {self.name} has {self.energy} % energy left.")

    def status(self):
        if self.energy < 40:
            print(f"{self.name} is currently {self.age} years old, and has {self.energy} % energy. {self.name} probably needs some sleep soon.")
        else:
            print(f"{self.name} is currently {self.age} years old, and has {self.energy} % energy. {self.name} wants to play!")

Stoffe = Dog('Stoffe', 'His', 7, 50)

Stoffe.birthday()
Stoffe.birthday()
Stoffe.bark()
Stoffe.sleep(2)
Stoffe.play(50)
Stoffe.play(50)

Stoffe.sleep(1)
Stoffe.sleep(5)
Stoffe.sleep(1)
Stoffe.sleep(4)
Stoffe.status()
Stoffe.play(40)
Stoffe.play(50)
Stoffe.status()

```

```python
class Person:
    def __init__(self, name):
        self.name = name
        self._age = 25

    def get_age(self):
        print(f"{self.name} är {self._age} år gammal.")

    def set_age(self, value):
        if value < 0:
            print("The age can't be negative.")
        elif value > 130:
            self._age = value
            print(f"The person {self.name} is {self._age} years old, and most likely really dead by now.")
        else: 
            self._age = value
            print(f"The persons age, {self.name}, is now {self._age} years.")
Anna = Person('Anna Jörgdotter')


Anna.get_age()
Anna.set_age(35)
Anna.set_age(-35)
Anna.set_age(437)
Anna.set_age(13)
```
